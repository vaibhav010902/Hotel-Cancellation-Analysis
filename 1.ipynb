```python
column = ['lead_time','arrival_date_week_number','arrival_date_day_of_month','stays_in_weekend_nights','stays_in_week_nights','distribution_channel','meal','is_repeated_guest','previous_cancellations','previous_bookings_not_canceled','booking_changes','deposit_type','company','agent','days_in_waiting_list','adr','required_car_parking_spaces','total_of_special_requests']
file.drop(columns=column,inplace=True)
file.head()
```

```python
file.isna().sum()
```

```python
file.children = file.children.replace(['NA',np.nan],0)
file.country = file.country.replace(np.nan,'Unknown')
```

```python
file.isna().sum()
```

```python
file['guest_type'] = ""
for index, (i, j, k) in enumerate(zip(file.adults, file.children, file.babies)):
    if i == 1 and j == 0 and k == 0:
        file.at[index, 'guest_type'] = "single"
    elif i == 2 and j == 0 and k == 0:
        file.at[index, 'guest_type'] = "couple"
    elif i > 1 and j > 0 and k > 0:
        file.at[index, 'guest_type'] = "family"
    else:
        file.at[index, 'guest_type'] = "family"
```

```python
file.head()
```

```python
file['room_status'] = ""
for index, (i,j) in enumerate(zip(file.reserved_room_type,file.assigned_room_type)):
    if(i==j):
        file.at[index,'room_status'] = "Desired"
    else:
        file.at[index,'room_status'] = "Undesired"
```

```python
file['no._of_guest']=""
for index, (i,j,k) in  enumerate(zip(file.adults, file.children, file.babies)):
    file.at[index, 'no._of_guest'] = i+j+k
```
